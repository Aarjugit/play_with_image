```python
# prompt: correct the above code
#this code will only for black and white image cz I took one matrix only
#it is having grayscale value not having any deviation

import cv2
import numpy as np
import matplotlib.pyplot as plt

# function to display image with matplotlib
# image is 3D matrix
def display_image(title, image):
    plt.title(title)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

# create a random image of size 28x28 with big diagonal line
random_image = np.zeros((28, 28), dtype=np.uint8)  # Create a 3-channel image
for i in range(28):
    #top left corner(0,0) to bottom right corner(27,27)
    random_image[i, i] = 255  # Draw a white diagonal line
    #bottom right (0,27) to top right(27,0)
    random_image[i,28-i-1] = 255
    random_image[i,14] = 255

display_image("Diagonal Line Image", random_image)
```

```python
#step 1: Load an image
image=cv2.imread('image.jpg')
display_image("Original Image", image)
```

```python
print(type(image))
print(image.shape)
#output will show it's a 3 chanel image and each chanel will have 800rows x 461columns shape
#3 chanels are the RGB chanels
```

```python
#step2: crop the image
#here y come first cz the notation of matrix is mxn (row(y)x column(x))
#y will be the range from 50 to 300 and x will be 100 to 400 range do croping
cropped_image=image[50:300, 100:400] #crop from (y=50, x=100) to (y=300, x=400)
display_image("Cropped Image", cropped_image)
```

```python
cropped_image=image[50:55, 100:105]
print(cropped_image)
```

```python
#Adjust the brightness and contrast
#Increase brightness by adding a constant values to all pixels
bright_image=cv2.convertScaleAbs(image, alpha=1, beta=50) #alpha is contrast ,beta is brightness
display_image("Brightened Image", bright_image)
#you can not add random numbers like this
bright_image2=(image + 50) % 255
display_image("Brightened image", bright_image2)
```

```python
#step4: Apply a filter -Black and white(Grayscale)
gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
display_image("Grayscale Image", gray_image)
```

```python
#you can also take one channel
img2=image[:, :, 0]
display_image("Red Channel Image", img2)
```
